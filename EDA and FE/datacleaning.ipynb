First of all lets straightaway drop the rows with name value as null, as we can not input any random value for it.

```python
df.dropna(subset='NAME', inplace=True)
```

Similarly dropping rows with unknown host name

```python
df.dropna(subset='host name', inplace=True)
```

As both country and country code represents United states only, lets just drop 'country' and fill the missing values in country code with US.

```python
df.drop(['country'], axis=1, inplace=True)
```

```python
#Replace 'nan' with 'US' in 'country code' where it is null
df['country code'] = df['country code'].where(df['country code'].notnull(), 'US', inplace=True)
```

Lets also drop this duplicates

```python
#Droping duplicates
df.drop_duplicates(inplace=True)
```

```python
df[df.duplicated()]
```

```python
df['host_identity_verified'].value_counts()
```

Lets replace nan values in 'host_identity_verified' with equal number of both verified and unverified as this both is already in almost equal numbers 

```python
#Lets replace 'nan' values in 'host_identity_verified' with equal number of 'unconfirmed' and 'confirmed' entries to maintain the balance in the dataset
num_nan = df['host_identity_verified'].isnull().sum()
num_unconfirmed = num_nan // 2
num_confirmed = num_nan - num_unconfirmed
df.loc[df['host_identity_verified'].isnull(), 'host_identity_verified'] = ['unconfirmed'] * num_unconfirmed + ['confirmed'] * num_confirmed
```

Replace 'confirmed' with 'verified'because 'confirmed' and 'verified' are essentially the same in the context of host identity verification, and we want to maintain consistency in our dataset.

```python
df['host_identity_verified'] = df['host_identity_verified'].replace('confirmed', 'verified')
```

```python
df['host_identity_verified'].value_counts()
```

```python
df.isnull().sum()
```

Dropping columns with null values in 'neighbourhood group' & 'neighbourhood' as their is as such no replacement or dummy value we can input

```python
# Lets drop rows that have 'nan' values in  'neighbourhood group' and 'neighbourhood' columns as they are important for our analysis and the number of such rows is very less
df.dropna(subset=['neighbourhood group', 'neighbourhood'], inplace=True)
```

## Now let's clean rest of the data in one go

```python
# Lets replace 'nan' values in house rules with 'No rules' to maintain the balance in the dataset
df['house_rules'] = df['house_rules'].where(df['house_rules'].notnull(), 'No rules', inplace=True)  
```

```python
#lets replace 'nan' values in license with 'No license' to maintain the balance in the dataset
df['license'] = df['license'].where(df['license'].notnull(), 'No license', inplace=True)
```

```python
#lets replace 'nan' values in 'Construction year' with the median value of the column
median_construction_year = df['Construction year'].median()
df['Construction year'] = df['Construction year'].where(df['Construction year'].notnull(), median_construction_year, inplace=True)
```

```python
#lets replace 'nan' values in 'lat' and 'long' with the average value of the respective columns
mean_lat = df['lat'].mean()
mean_long = df['long'].mean()
df['lat'] = df['lat'].where(df['lat'].notnull(), mean_lat, inplace=True)
df['long'] = df['long'].where(df['long'].notnull(), mean_long, inplace=True)    
```

```python
#lets replace 'nan' values in 'instant_bookable' with randomly assigned 'True' and 'False' values to maintain the balance in the dataset
num_nan_instant_bookable = df['instant_bookable'].isnull().sum()
num_true = num_nan_instant_bookable // 2
num_false = num_nan_instant_bookable - num_true
df.loc[df['instant_bookable'].isnull(), 'instant_bookable'] = [True] * num_true + [False] * num_false   
```

```python
#lets replace 'nan' values in 'cancellation_policy' with randomly assigned 'flexible', 'moderate' and 'strict' values to maintain the balance in the dataset
num_nan_cancellation_policy = df['cancellation_policy'].isnull().sum()
df.loc[df['cancellation_policy'].isnull(), 'cancellation_policy'] = np.random.choice(['flexible', 'moderate', 'strict'], size=num_nan_cancellation_policy)  
```

```python
df['price'].unique()
```

```python
#lets replace 'nan' values in 'price' with the average value of the column
mean_price = df['price'].mean()
df['price'] = df['price'].where(df['price'].notnull(), mean_price, inplace=True)
```

```python
#lets first covert the 'price' column to numeric values by removing the '$' sign and then replacing 'nan' values in 'price' with the average value of the column
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
mean_price = df['price'].mean()
df['price'] = df['price'].where(df['price'].notnull(), mean_price, inplace=True)
```

```python
df['service fee'].unique() 
```

```python
#lets first convert the 'service fee' column to numeric values and then replace 'nan' values with the average value of the column
df['service fee'] = df['service fee'].replace('[\$,]', '', regex=True).astype(float)
mean_service_fee = df['service fee'].mean()
df['service fee'] = df['service fee'].where(df['service fee'].notnull(), mean_service_fee, inplace=True)
```

```python
#lets replace 'nan' values in 'reviews per month' with the average value of the column
mean_reviews_per_month = df['reviews per month'].mean()
df['reviews per month'] = df['reviews per month'].where(df['reviews per month'].notnull(), mean_reviews_per_month, inplace=True)
```

```python
#lets replace 'nan' values in 'review rate number' with the average value of the column
mean_review_rate_number = df['review rate number'].mean()
df['review rate number'] = df['review rate number'].where(df['review rate number'].notnull(), mean_review_rate_number, inplace=True)
```

```python
#lets replace 'nan' values in 'calculated host listings count' with the mode value of the column
mode_calculated_host_listings_count = df['calculated host listings count'].mode()[0]
df['calculated host listings count'] = df['calculated host listings count'].where(df['calculated host listings count'].notnull(), mode_calculated_host_listings_count, inplace=True)
```

```python
# Lets replace 'nan' values in 'last review' with the most recent date in the column to maintain the balance in the dataset
most_recent_date = df['last review'].max()
df['last review'] = df['last review'].where(df['last review'].notnull(), most_recent_date, inplace=True)
```

```python
# number of reviews column has some outliers. Lets replace those outliers with the median value of the column to maintain the balance in the dataset
median_number_of_reviews = df['number of reviews'].median()
df['number of reviews'] = df['number of reviews'].where(df['number of reviews'] <= 100, median_number_of_reviews, inplace=True)
```

```python
# availability 365 column has some outliers. Lets replace those outliers with the median value of the column to maintain the balance in the dataset
median_availability_365 = df['availability 365'].median()
df['availability 365'] = df['availability 365'].where(df['availability 365'] <= 200, median_availability_365, inplace=True) 
```

```python
# Minimum nights column has some outliers. Lets replace those outliers with the median value of the column to maintain the balance in the dataset
median_minimum_nights = df['minimum nights'].median()
df['minimum nights'] = df['minimum nights'].where(df['minimum nights'] <= 30, median_minimum_nights, inplace=True)
```

```python
# Still minimum nights column has some outliers. Lets replace those outliers with the median value of the column to maintain the balance in the dataset
median_minimum_nights = df['minimum nights'].median()
df['minimum nights'] = df['minimum nights'].where(df['minimum nights'] <= 14, median_minimum_nights, inplace=True)
```

```python
df['minimum nights'].value_counts()
```

Why are minimum night column having negative values? Are those values correct or they are data entry errors? We need to investigate this further to understand the reason behind these negative values in minimum nights column.


```python
# Replace negative values in 'minimum nights' column with the median value of the column to maintain the balance in the dataset
median_minimum_nights = df['minimum nights'].median()
df['minimum nights'] = df['minimum nights'].where(df['minimum nights'] >= 0, median_minimum_nights, inplace=True)   
```
